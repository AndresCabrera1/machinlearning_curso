# 2. importacion de la data

```python
# importacion tradicional de la data
data_set_clas_1 = 'data_clas_1.csv'
data_clas_1 = pd.read_csv('../Modulo_1/dataset/'+data_set_clas_1)
data_clas_1.head()
```

```python
# importacion tradicional de la data : funciones 
def importData(path,fileName):
    datos = pd.read_csv(path+fileName)
    return datos

example = importData('../Modulo_1/dataset/',data_set_clas_1)
example.head()
```

```python
#importar datos en un JS
data_set_clas_1_js = 'data.js'
def importDataJs(path,fileName):
    datos = pd.read_json(path+fileName)
    return datos

example = importDataJs('./dataset/',data_set_clas_1_js)
example.head()
```

```python
# importar data desde archivo en la nube
# el parametro sep define la separacion de casa selda del archivo en este caso esta separado por ';'

def importNube(pathUrl, sep ):
    df = pd.read_csv(pathUrl,sep=sep)
    return df

fileNube= importNube('https://query.data.world/s/lzwxkoqy7xwsuf255pzz2py7usrt6u',';')
fileNube.head()
```

```python
# importar data desde archivo en formato excel xlsx
data_set_clas_1_xls = 'Productos.xlsx'
def importExcel(path,fileName,col1, col2):
    ex= pd.read_excel(path+fileName)
    ex2 = pd.DataFrame(ex,columns=[col1,col2])
    return ex2

data_excel = importExcel('./dataset/',data_set_clas_1_xls,'producto','precio')
data_excel.head()
```

```python
#importar multiples hojas de excel con el mismo encabezado

def importExcelMul(url,fileName):
    df = pd.concat(pd.read_excel(url+fileName,sheet_name=None), ignore_index=True)
    return df

dataConc = importExcelMul('https://github.com/chris1610/pbpython/raw/master/data/','2018_Sales_Total_Tabs.xlsx')

dataConc.head()
```
