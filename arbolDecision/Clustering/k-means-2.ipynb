Diccionario de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
# Se cargan los datos
dataframe = pd.read_csv("data_set/analisis.csv")
dataframe.head()
```

```python
# Se muestran las estadísticas
dataframe.describe()
```

```python
# Se muestran los registros para cada categoría
dataframe.groupby('categoria').size()
```

```python
# Se genera un histograma para ver la distribución de los datos
dataframe.drop(['categoria'],1).hist()
plt.show()
```

```python
#Definimos las entradas
X = np.array(dataframe[["op","ex","ag"]])
y = np.array(dataframe['categoria'])
```

```python
#Obtenemos el valor de K
Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X).score(X) for i in range(len(kmeans))]
score
plt.plot(Nc,score, marker = 'o')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
#K-Means
# Se define el modelo
kmeans = KMeans(n_clusters=5).fit(X)
# Se idnetifican los centroides
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
# Se crean los grupos
f1 = dataframe['op'].values
f2 = dataframe['ex'].values
labels = kmeans.predict(X)

C = kmeans.cluster_centers_
colores=['red','green','blue','cyan','yellow']
asignar=[]
for row in labels:
    asignar.append(colores[row])

plt.scatter(f1, f2, c=asignar, s=70)
plt.scatter(C[:, 0], C[:, 1], marker='*', c=colores, s=1000)
plt.show()
```

```python
#Observar clustering por grupos
# Se observa la cantidad de elementos por cluster
copy = pd.DataFrame()
copy['usuario']=dataframe['usuario'].values
copy['categoria']=dataframe['categoria'].values
copy['label'] = labels
cantidadGrupo = pd.DataFrame()
cantidadGrupo['color']=colores
cantidadGrupo['cantidad']=copy.groupby('label').size()
cantidadGrupo
```
